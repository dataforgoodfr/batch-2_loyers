# Get front page information

```python
def get_front(url):
    with urllib.request.urlopen(url) as url:
        s = url.read()
    soup = BeautifulSoup(s)
    offers = soup.findAll('div', class_="header-annonce")
    url_list = []
    id_list = []
    title_list = []
    price_list = []
    location_list = []
    release_list = []
    for i in range(0, 10):
        try:
            url_list.append('http://www.pap.fr'+offers[i].a['href'])
        except Exception as e:
            url_list.append('')
        try:
            id_list.append(offers[i].a['name'])
        except Exception as e:
            id_list.append('')
        try:
            title_list.append(offers[i].findAll('span', class_='desc')[0].get_text())
        except Exception as e:
            title_list.append('')
        try:
            price_list.append(offers[i].findAll('span', class_='prix')[0].get_text())
        except Exception as e:
            price_list.append('') 
        try:
            release_list.append(offers[i].findAll('span', class_='date')[0].get_text().lstrip().rstrip())
        except Exception as e:
            release_list.append('')
    return [url_list, id_list, title_list, price_list, release_list]
```

```python
start = time.time()
a = get_front(pap)
print(time.time() - start)
```

```python
def get_page_info(url):
    front = get_front(url)
    description_list = []
    location_inside_list = []
    room_list = []
    bedroom_list = []
    superficy_list = []
    metro_list = []
    latitude_list = []
    longitude_list = []
#     ges_list = []
#     energy_list = [] 
    for i in front[0]:
        with urllib.request.urlopen(i) as url:
            s = url.read()
        soup = BeautifulSoup(s)
        
        try:
            latitude_list.append(soup.findAll('div', class_="map-annonce-adresse")[0]['data-map-lat'])
        except Exception as e:
            latitude_list.append('')
            
        try:
            longitude_list.append(soup.findAll('div', class_="map-annonce-adresse")[0]['data-map-lng'])
        except Exception as e:
            longitude_list.append('')
        
        try:
            bedroom_list.append(re.findall('[0-9]', str(soup.findAll('div', class_="footer-descriptif clearfix")[0]))[1])
        except Exception as e:
            bedroom_list.append('')    
        try:
            description_list.append(soup.p.get_text())
        except Exception as e:
            description_list.append('')
        try:
            location_inside_list.append(soup.h2.get_text())
        except Exception as e:
            location_list.append('')
        try:
            room_list.append(re.findall('[0-9]', str(soup.findAll('div', class_="footer-descriptif clearfix")[0].li))[0])
        except Exception as e:
            room_list.append('')
            
        try:
            metro = soup.findAll('div', class_='metro')[0].findAll('span', class_='label')
            metro = [metro[i].get_text() for i in range(0, len(metro))]
            metro_list.append(metro)
        except Exception as e:
            metro_list.append('')

        try:
            superficy_list.append(re.findall('[0-9]+', str(soup.findAll('li', class_='last')[0]))[0])
        except Exception as e:
            superficy_list.append('')
#         try:
#             ges_list.append(infos[6].a.get_text())
#         except Exception as e:
#             ges_list.append('')
#         try:
#             energy_list.append(infos[7].a.get_text())
#         except Exception as e:
#             ges_list.append('')
            
    a = pd.DataFrame(front).transpose()
    b = [description_list, location_inside_list, room_list, bedroom_list, superficy_list, metro_list, latitude_list, longitude_list]
    b = pd.DataFrame(b).transpose()
    a.columns = ['url', 'id', 'title', 'price', 'release']
    b.columns = ['description', 'location', 'rooms', 'bedrooms', 'superficy', 'metro', 'latitude', 'longitude']
    result = pd.concat([a, b], axis=1)
    
    return result
```

```python
start = time.time()
a = get_page_info(pap)
print(time.time() - start)
```

# And now let's build final scraping function

```python
list_pages = ['http://www.pap.fr/annonce/locations-appartement-paris-75-g439-%i' %i for i in range(1, 5)]
list_pages
```

```python
def get_n_pages(N):
    list_pages = ['http://www.pap.fr/annonce/locations-appartement-paris-75-g439-%i' %i for i in range(1, N+1)]
    a = get_page_info(list_pages[0])
    
    for i in list_pages[1:]:
        ind = list_pages[1:].index(i)
        b = get_page_info(i)
        frames = [a, b]
        a = pd.concat(frames).reset_index()
        a = a.drop('index', 1)
        print('Page %i has been scrapped' %ind)
        time.sleep(4)
    return a
```

```python
start = time.time()
d = get_n_pages(100)
print(time.time() - start)
```
